#### Question 1 (True risk / Empirical risk):


True risk $R(f)=\mathbb{E}_D[L(f(x), y)]$.  i.e., expected loss between prediction and "true data distribution"  
Empirical risk $\hat{R}(f)=\frac{1}{N} \sum_{i=1}^N L\left(f\left(x_i\right), y_i\right) .$  i.e., average loss between prediction and data(samples)  
The core difference between them is we almost never know the real distribution of data (there is no point of doing machine learning if we know it), thus we can't calculate the true risk.Empirical risk is an approximation using information we know to make it pratical.

```python
##------------------task 1----------------------------------
class Net(nn.Module):
    def __init__(self, dim, nclass, width, depth):
        super(Net, self).__init__()
        layers = [nn.Flatten()] #initialize layers and add flatten layer
        #----set input layer and and first hidden layer
        layers.append(nn.Linear(dim, width))
        layers.append(nn.ReLU())
        #-----set hidden layer
        for _ in range(1, depth):
            layers.append(nn.Linear(width, width))
            layers.append(nn.ReLU())
        #------connect last hidden layer with output layer
        layers.append(nn.Linear(width, nclass))
        self.model = nn.Sequential(*layers)
    
    def forward(self, x):
        return self.model(x)

##------------------task 2----------------------------------
from torch.utils.data import DataLoader
def loading_data(batch_size, train_set, test_set):

    trainloader = DataLoader(train_set, batch_size=batch_size, shuffle=True)
    
    testloader = DataLoader(test_set, batch_size=batch_size, shuffle=False)
    
    return trainloader, testloader            
```

```python
#------------------------task3-------------------
def train_epoch(trainloader, net, optimizer, criterion):
    net.train() 
    total_loss = 0.0
    for data in trainloader:
        # load data from loader
        inputs, labels = data
        
        # intialize gradient
        optimizer.zero_grad()
        
        #pass data
        outputs = net(inputs)
        
        # calculate loss
        loss = criterion(outputs, labels)
        
        # calculate gradient based on loss
        loss.backward()
        
        # update weights
        optimizer.step()
        
        # add loss
        total_loss += loss.item()
    
    # calculate average loss
    epoch_loss = total_loss / len(trainloader)
    
    return epoch_loss

#------------------------task4-------------------
def test_epoch(testloader, net, criterion):
    net.eval()  # set the model in evaluation mode
    total_loss = 0.0
    total_error = 0
    
    with torch.no_grad():  
        for data in testloader:
            inputs, labels = data
            outputs = net(inputs)
            
            # calculate loss
            loss = criterion(outputs, labels)
            total_loss += loss.item()
            
            # count error
            _, predicted = torch.max(outputs.data, 1)
            total_error += (predicted != labels).sum().item()
    
    # calculate average loss 
    average_loss = total_loss / len(testloader)
    
    return average_loss, total_error
```

```python

def run_training(train_set, test_set, batch_size=64, dim=28*28, nclass=10, width=512, depth=4, learning_rate=0.001, num_epochs=15):
  

    trainloader, testloader = loading_data(batch_size, train_set, test_set)


    net = Net(dim, nclass, width, depth)

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(net.parameters(), lr=learning_rate)

    for epoch in range(num_epochs):
        train_loss = train_epoch(trainloader, net, optimizer, criterion)
        test_loss, test_error = test_epoch(testloader, net, criterion)

        print(f'Epoch: {epoch+1:03} | Train Loss: {train_loss:.4f} | Test Loss: {test_loss:.4f} | Test Error: {test_error:.4f}')
    
    return train_loss,test_loss

#run_training(train_set_mnist,test_set_mnist)
```

### Task 1: Proper weight scaling

Consider the $var[\sum_{j=1}^N w_{ij}g^{1}_j(x)] = E((\sum_{j=1}^N w_{ij}g^{1}_j(x))^2) - E((\sum_{j=1}^N w_{ij}g^{1}_j(x)))^2$  
Since w is iid with zero mean, $E(w_{ij}g_j(x)) = 0$, $E((\sum_{j=1}^N w_{ij}g^{1}_j(x)))^2 = 0$, then  
$var[\sum_{j=1}^N w_{ij}g^{1}_j(x)] = E[(\sum_{j=1}^N w_{ij}g^{1}_j(x))^2] = NE(w^2)E(g(x)^2) = NC_w^{(l)}E(g(x)^2) $ g is determined by previous weight and bias, or data, independent to w.
To avoid infinity variance which can lead to problem such as gradient explosion, use the scalling can make the value fixed to a desired value $\sigma_w$.

### Task 2: Derive the GP relation for a single hidden layer

vector form  
$\mathbf{f}_i^{(2)} = \mathbf{w}^{(2)}\mathbf{g}^{(1)} + \mathbf{b}^{(2)}$
where $\mathbf{w}^{(2)}$ is $N_2 \times N_1$ matrix storing weights and $ \mathbf{b}^{(2)}$ is vector of biases for the second layer, $\mathbf{g}^{(1)}$ is outputs from the first layer.  
First checking expectation, we know $E[\mathbf{b}^{(2)}] = 0$, $E[\mathbf{W}^{(2)}\mathbf{g}^{(1)}] = 0 $ since  $E[\mathbf{W}^{(2)}] = 0$.  
Now check variance, $var(\mathbf{f}_i^{(2)}) = \sigma_b^2 + var[\mathbf{W}^{(2)}\mathbf{g}^{(1)}] =  \sigma_b^2 + \sigma_w^2 E[\mathbf{g}^{(1)}(x)^2]$   
apply central limit theorm, we have $\mathbf{f}^{(2)} \sim N(0, \sigma_b^2 + \sigma_w^2 E[\mathbf{g}^{(1)}(x)^2])$  
suppose we have data set $x^1,x^2,x^3,\cdots,x^n$
the prior joint distribution converge to multivariate Guassian distribution with 0 mean and covariance between two datas $x^p,x^q$ to be $\sigma_b^2 + \sigma_w^2E(\phi(f^{(1)}(x^p))\phi(f^{(1)}(x^q)))$ also known as Guassian process.

### Task 3: Why in succession

The need to apply the limit in succession to make sure each layer we analyzing are in Guassian process. This is infinity first layer implies guassian process in nodes of second layer, and infity second layer implies GP is third layer and so forth.

### Task 4: Derive the GP relation for multiple hidden layers

Suppose $f^{l-1}$ is a GP hence $g_j^{l-1}$ are iid, and we want to check $f^l = W^{(l)}g^{(l-1)}(x) + b^{(l)} $   
similary, mean of $f^l$ is zero, and variance is $\sigma_b^2 + \sigma_w^2E(g^{(l-1)}(x)^2)$   
there are serval conditions for CLT, firstly we need variable are iid, we know w,b,g(x) are all iid.s hence $f^{(l)}$ is iid. Secondly we need finite variance. $\sigma_b, \sigma_w$ are constance and  $E(g^{(l-1)}(x)^2)$ is bonded since the activation is bonded. Hence we have finite varinace. Therfore CLT can be applied in this case.
with the similar idea, the covariance of sample $x^p,x^q$, or the kernel of this layer is

```python
#--------------task1
class Kernel:
    def __init__(self, w, b, L, dim):
        self.w = w
        self.b = b
        self.L = L
        self.dim = dim

    def K_0(self, X1, X2):
        """Compute the base kernel matrix."""
        return self.b + self.w * torch.matmul(X1, X2.T) / self.dim
    
    def compute_arc_cosine_kernel(self, K_curr, K_diag_X1, K_diag_X2):
        """Helper function to compute the arc cosine kernel."""
        theta = torch.acos(torch.clamp(K_curr / torch.sqrt(torch.ger(K_diag_X1, K_diag_X2)), -1, 1))
        return self.b + self.w * torch.sqrt(torch.ger(K_diag_X1, K_diag_X2)) * \
               (torch.sin(theta) + (torch.pi - theta) * torch.cos(theta)) / (2 * torch.pi)

    def NN_K(self, X1, X2):
        """Compute the neural network kernel matrix using the arc-cosine kernel."""
        # Initial kernel matrices
        K = self.K_0(X1, X2)
        K_X1X1 = self.K_0(X1, X1)
        K_X2X2 = self.K_0(X2, X2)

        for _ in range(self.L - 1):
            # Compute diagonal elements
            diag_K_X1X1 = torch.diag(K_X1X1)
            diag_K_X2X2 = torch.diag(K_X2X2)

            # Update kernels using the arc-cosine kernel formula
            K = self.compute_arc_cosine_kernel(K, diag_K_X1X1, diag_K_X2X2)
            K_X1X1 = self.compute_arc_cosine_kernel(K_X1X1, diag_K_X1X1, diag_K_X1X1)
            K_X2X2 = self.compute_arc_cosine_kernel(K_X2X2, diag_K_X2X2, diag_K_X2X2)

        return K

#-----------task2
class GP:
    def __init__(self, kernel, X, y, noise):
        self.kernel = kernel  # The kernel function
        self.X = X  # Training input data
        self.y = y  # Training output data
        self.noise = noise  # Noise term
        # Precompute the kernel matrix with noise term added
        self.temp = self.kernel.NN_K(X, X) + noise * torch.eye(X.size(0))
    
    def replace_tensor_values(self,tensor):
        tensor[tensor > 0] = 0.5
        tensor[tensor < 0] = -0.5
        return tensor
    
    def predict(self, X_t):
        """Predict mean and variance for the test inputs X_t."""
        # Compute covariance between test and training data, and test data alone
        K_s = self.kernel.NN_K(X_t, self.X)
        K_ss = self.kernel.NN_K(X_t, X_t)
        
        # Compute predictive mean
        pred_mean = self.replace_tensor_values(torch.matmul(K_s, torch.linalg.solve(self.temp, self.y)))
        
        # Compute predictive variance
        pred_var = K_ss - torch.matmul(K_s, torch.linalg.solve(self.temp, K_s.T))

        self.prediction = pred_mean
        self.var = pred_var
        return pred_mean, pred_var
    
    def accuracy(self, y_true):
        """Calculate the accuracy of the predictions."""
        if not hasattr(self, 'prediction'):
            raise ValueError("You need to call predict method before calculating accuracy.")
        # Ensure the prediction and true labels are on the same device and have the same shape
        y_true = y_true.to(self.prediction.device)
        correct_predictions = (self.prediction == y_true).float().mean()
        return correct_predictions.item()
    
    def get_extreme_variance_indices(self):
        if not hasattr(self, 'var'):
            raise ValueError("You need to call predict method before accessing variances.")
        var_diagonal = self.var.diag()  
        
        
        max_var_indices = torch.topk(var_diagonal, 2).indices  
        min_var_indices = torch.topk(var_diagonal, 2, largest=False).indices 
        
        return max_var_indices.tolist(), min_var_indices.tolist()
    
    

sig_b = [i*0.02 for i in range(10)]
sig_w = [i*0.02 for i in range(10)]
L = [i for i in range(1,11)]

print(y_test.shape)    
kernel = Kernel(0.1, 0.1, 1, 3072)
gp_model = GP(kernel, X_train, y_train, noise=0.1)
pred_mean, pred_var = gp_model.predict(X_test)
accuracy = gp_model.accuracy(y_test)
print(f"Test Accuracy: {accuracy}")
pred_var.shape
```

```python
from tqdm import tqdm
#------------task 3
def evaluate_model(X_train, y_train, X_test, y_test, b, w, l):
    kernel = Kernel(w, b, l, 3072)
    gp_model = GP(kernel, X_train, y_train, noise=0.01)
    _, _ = gp_model.predict(X_test) 
    return gp_model.accuracy(y_test)

def find_best_sigb_sigw_combinations(X_train, y_train, X_test, y_test):
    sig_b = [i * 0.5 for i in range(10)]
    sig_w = [i * 0.5 for i in range(10)]
    L_values = [i for i in range(1, 11)]
    
    best_combinations = [] 
    

    for L in tqdm(L_values, desc="Optimizing over L"):
        best_accuracy = 0
        best_sigb = 0
        best_sigw = 0
        
        for b in sig_b:
            for w in sig_w:

                accuracy = evaluate_model(X_train, y_train, X_test, y_test, b, w, L)
                
                if accuracy > best_accuracy:
                    best_accuracy = accuracy
                    best_sigb = b
                    best_sigw = w
        
        best_combinations.append([best_accuracy, best_sigb, best_sigw])
    
    return best_combinations

res = find_best_sigb_sigw_combinations(X_train, y_train, X_test, y_test)
res 
```
